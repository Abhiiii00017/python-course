```python
def clean_price_coloumn(price):
    if isinstance(price,str):
        price = price.replace(',','')
        return float(price)
    return price
```

```python
clean_price_coloumn('236,546')
```

```python
df['price'].apply(clean_price_coloumn)
```

```python
df['clean_price']= df['price'].apply(clean_price_coloumn)
df
```

```python
def clean_rating_coloumn(rating):
    if isinstance(rating,str):
        if len(rating)>3:
            return np.nan
        if rating.count('.')==1:
            return float(rating)
        if rating.isnumeric():
            return int(rating)
        return np.nan
    return rating
```

```python
clean_rating_coloumn('1 stock left only')
```

```python
df ['clean_rating']= df['rating'].apply(clean_rating_coloumn)
```

```python
def clean_oprice_coloumn(oprice):
    if isinstance(oprice,str):
        oprice = oprice.replace('₹','')
        oprice = oprice.replace(',','')
        return float(oprice)
    return oprice
```

```python
clean_oprice_coloumn('₹456')
```

```python
df['oprice'].apply(clean_oprice_coloumn)
```

```python
df ['clean_oprice'] = df['oprice'].apply(clean_oprice_coloumn)
df
```
